```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('matches.csv')
```

Q1 Evaluate the freq of most man of the match award

```python
man_of_match_freq = df['player_of_match'].value_counts()
print(man_of_match_freq)
```

Q2 Evaluate the top 10 players with most man of the match award

```python
top_10_players = man_of_match_freq.head(10)
print(top_10_players)
```

Q3 Extract the top 5 players with most man of the match award

```python
top_5_players = man_of_match_freq.head(5)
print(top_5_players)
```

Q4 Draw bar plot of top 5 most man of the match award winning players

```python
plt.figure(figsize=(10, 6))
top_5_players.plot(kind='bar', color='skyblue')
plt.title('Top 5 Players with Most Man of the Match Awards')
plt.xlabel('Player')
plt.ylabel('Number of Awards')
plt.grid(True)
plt.show()

```

Q5 Draw violen plot of top 5 player most man of the match winning player

```python
plt.figure(figsize=(10, 6))
plt.violinplot(top_5_players)
plt.title('Violin Plot of Top 5 Players with Most Man of the Match Awards')
plt.grid(True)
plt.show()
```
